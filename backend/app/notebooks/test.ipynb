```python
import pandas as pd
```

```python
df=pd.read_csv(r'D:\EDrivebackup\brocamp\techTest\fastapi-ml-dashboard\backend\data\Churn_Modelling.csv')
df.head()
```

```python
schema = {
    "row_count": len(df),
    "column_count": df.shape[1],
    "columns": []
}

for col in df.columns:
    col_data = df[col]
    dtype = str(col_data.dtype)
    nunique = col_data.nunique()
    null_pct = col_data.isnull().mean() * 100
    is_constant = nunique == 1
    is_high_cardinality = nunique > 0.8 * len(df)

    # Detect inferred type (string, number, bool, datetime)
    if pd.api.types.is_numeric_dtype(col_data):
        inferred_type = "numerical"
    elif pd.api.types.is_bool_dtype(col_data):
        inferred_type = "boolean"
    elif pd.api.types.is_datetime64_any_dtype(col_data):
        inferred_type = "datetime"
    else:
        inferred_type = "categorical"

    schema["columns"].append({
        "name": col,
        "dtype": dtype,
        "inferred_type": inferred_type,
        "null_percentage": round(null_pct, 2),
        "unique_count": int(nunique),
        "is_constant": is_constant,
        "is_high_cardinality": is_high_cardinality,
        "sample_values": col_data.dropna().unique()[:5].tolist()  # For UI preview
    })
```
